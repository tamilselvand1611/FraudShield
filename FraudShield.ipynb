```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("creditcard.csv")
data
```

```python
print("\nDataset Shape:")
print(data.shape)
print("\nMissing Values:")
print(data.isnull().sum())
print("\nDuplicate Values:")
print(data.duplicated().sum())
data=data.drop_duplicates()
print("\nDataset Shape After Removing Duplicates:")
print(data.shape)
```

```python
print(data.head())
print(data.columns)
```

```python
print("\nClass Distribution:")
print(data["Class"].value_counts())
```

```python
X=data[["Time", "Amount", "V1", "V2", "V3"]]
y=data["Class"]
```

```python
print(X.head())
print(y.head())
```

```python
model=GaussianNB()
model.fit(X, y)
```

```python
pred=model.predict(X)
print(pred)
```

```python
acc=accuracy_score(y, pred)
print("Accuracy Score:",acc*100,"%")
```

```python
cm=confusion_matrix(y, pred)
print("Confusion Matrix:")
print(cm)
```

```python
time=float(input("Enter Time:"))
amount=float(input("Enter Amount:"))
v1=float(input("Enter V1:"))
v2=float(input("Enter V2:"))
v3=float(input("Enter V3:"))
test_transaction=pd.DataFrame({"Time":[time], "Amount":[amount], "V1":[v1], "V2":[v2], "V3":[v3]})
```

```python
prediction=model.predict(test_transaction)
print("Predicted Transaction:",prediction[0])
```

```python
if prediction[0]==1:
    print("Fraudulent Transaction Detected")
else:
    print("Legitimate Transaction")
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=["Legitimate", "Fraud"], yticklabels=["Legitimate", "Fraud"])
plt.title("FraudShield Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
